#### 22)Depict Market and Type wise sum of Budget Profit on a bar chart

```python
Budget=df.groupby(['Type','Market'])['Budget Profit'].sum().reset_index()
pivote_budget=Budget.pivot(index='Market',columns='Type',values='Budget Profit')
print(pivote_budget)
```

```python
pivote_budget.plot(kind='bar')
plt.show()
```

#### Depict Product Type and State wise mean of Sales on a area chart

```python
sales=df.groupby(['Product Type','State'])['Sales'].mean().reset_index()
```

```python
pivote_data=sales.pivot(index='State',columns='Product Type',values='Sales')
print(pivote_data)
```

```python
pivote_data.plot(kind='area')
plt.show()
```

#### Depict Product and Type wise sum of Budget COGS on a line Chart

```python
Budget_COGS=df.groupby(['Type','Product'])['Budget COGS'].sum().reset_index()
pivote_budgetCOGS=Budget_COGS.pivot(index='Product',columns='Type',values='Budget COGS')
print(pivote_budget)
```

```python
pivote_budgetCOGS.plot(kind='line')
plt.xticks(rotation=45)
plt.show()
```

#### Depict Product and Type wise sum of Budget Profit and Budget Sales on a line Char

```python
Budget=df.groupby(['Type','Product'])[['Budget Profit','Budget Sales']].sum().reset_index()
pivote_budget_profit=Budget.pivot(index='Product',columns='Type',values='Budget Profit')
pivote_budget_sales=Budget.pivot(index='Product',columns='Type',values='Budget Sales')
print(pivote_budget_profit)
print(pivote_budget_sales)
```

```python
pivote_budget_profit.plot(kind='line')
plt.xticks(rotation=45)

pivote_budget_sales.plot(kind='line')
plt.xticks(rotation=45)

plt.show()
```

#### Depict Market and Product Type based mean of Sales on abar chart and on a stacked bar chart separately
